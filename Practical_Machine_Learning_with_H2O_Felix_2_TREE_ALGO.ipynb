```python
import h2o
```

```python
h2o.init()
```

```python
url = "http://h2o-public-test-data.s3.amazonaws.com/smalldata/iris/iris_wheader.csv"
iris = h2o.import_file(url)
type(iris) #felix
train, test = iris.split_frame([0.8])

#felix
print("type of train object is ", type(train))
print("training data dimensions ", train.shape)

#felix
print("type of test object is ", type(test))
print("test data dimensions ", test.shape)

train.summary()
```

```python
train.nrows
```

```python
test.nrows
```

```python
all_cols = train.col_names
print(all_cols)
```

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator
```

```python
x_cols = all_cols[:-1]
print(x_cols)
```

```python
y_col = all_cols[len(all_cols)-1]
print(y_col)
```

# Random Forest

```python
mRF = H2ORandomForestEstimator()
mRF.train(x_cols, y_col, train)
```

```python
mRF
```

```python
pred = mRF.predict(test)
print(pred.head)
```

```python
mRF.model_performance(test)
```

# GBM

```python
from h2o.estimators.gbm import H2OGradientBoostingEstimator
```

```python
mGBM = H2OGradientBoostingEstimator()
```

```python
mGBM.train(x_cols, y_col, train)
```

```python
mGBM ### IT GOT A PERFECT SCORE
```

```python
pred = mGBM.predict(test)
```

```python
mGBM.model_performance(test)
```
